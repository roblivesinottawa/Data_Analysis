## Libraries

```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default

from matplotlib.ticker import PercentFormatter #converts values into percentage format
```

## Bar Chart

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
df_used_cars
```

```python
plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"])
plt.show()
```

```python
# how to fix the overlapping labels
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"])
plt.show()
```

```python
# another way to improve visibility is to rotate the labels

plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"])
plt.xticks(rotation=45)
plt.show()
```

```python
# choosing a color
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"], color="rg")
plt.show()
# rgbwymc = red, green, blue, white, yellow, magenta, cyan
```

```python
# to integrate a title
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"], color="midnightblue")
plt.xticks(rotation=45, fontsize=13)
plt.yticks(fontsize=13)
plt.title("Cars Listings By Brand", fontsize=16, fontweight="bold")
plt.ylabel("Number of Listings", fontsize=13)
plt.show()
```

```python
# how to save and export image
plt.savefig("used_cars_bar.png")
```

## Stacked Area Chart

```python
# an area chart is used to display quantitative data for three or more numerical values
df_fuel_engine_type = pd.read_csv('stacked_area_chart_data.csv')
df_fuel_engine_type
```

```python
# define colors(4)
colors = ['#011638', '#7E2987', 'EF2026']
# define labels(4)
labels = ['Diesel', 'Petrol', 'Gas']
# using seaborn, create a white background
sns.set_style('white')
# increase the size
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_type['Year'],
                df_fuel_engine_type['Diesel'],
                df_fuel_engine_type['Petrol'],
                df_fuel_engine_type['Gas'],
                colors = colors,
                edgecolor = 'none',
                
                )
# include the years and make sure the labels don't overlap
plt.xticks(df_fuel_engine_type['Year'], rotation=45)
# include the labels
plt.legend(labels = labels, loc='upper left')
# create a Y-AXIS label
plt.ylabel('Number of Cars', fontsize=13, fontweight='bold')
# include title
plt.title('Popularity of Engine Fuel Types (1982-2016)', fontsize=14, fontweight='bold')
# despine allows us to remove the borders
sns.despine(left=True, top=True)
plt.show()
```

## Line Chart

```python
df_spx_ftse_00_10 = pd.read_csv("line_chart_data.csv")
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10.size
```

```python
# convert 'Date' values to 'datetime' format
df_spx_ftse_00_10['new_date']  = pd.to_datetime(df_spx_ftse_00_10['Date'])
```

```python
df_spx_ftse_00_10['new_date']
```

```python
labels = ['S&P 500', 'FTSE 100']
plt.figure(figsize=(20,8))
plt.plot(df_spx_ftse_00_10['new_date'], df_spx_ftse_00_10['GSPC500'])
plt.plot(df_spx_ftse_00_10['new_date'], df_spx_ftse_00_10['FTSE100'])
plt.title("S&P Vs FTSE Returns (2000-2010)", fontsize=14, fontweight='bold')
plt.ylabel('Returns', fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=14, fontweight='bold')
plt.legend(labels = labels, fontsize='x-large')
plt.show()
```

```python
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10['new_date'] >= '2008-07-01') &
                                        (df_spx_ftse_00_10['new_date'] <= '2008-12-31')]
```

```python
df_spx_ftse_H2_08
```

```python
# labels = ['S&P 500', 'FTSE 100']
plt.figure(figsize=(20,8))
plt.plot(df_spx_ftse_H2_08['new_date'], df_spx_ftse_H2_08['GSPC500'], color = 'midnightblue')
plt.plot(df_spx_ftse_H2_08['new_date'], df_spx_ftse_H2_08['FTSE100'], color = 'red')
plt.title("S&P Vs FTSE Returns (H2 2008)", fontsize=14, fontweight='bold')
plt.ylabel('Returns', fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=14, fontweight='bold')
plt.legend(labels = labels, fontsize='x-large')
plt.show()
```

## Histogram

```python
df_real_estate = pd.read_csv('histogram_data.csv')
df_real_estate
```

```python
sns.set_style('white')
plt.figure(figsize=(8,6))
plt.hist(df_real_estate['Price'], bins=14, color='#108A99')
plt.title('Distribution of Real Estate Prices', fontsize=14, fontweight='bold')
# plt.xticks()
plt.xlabel("Price in (000's $)")
plt.ylabel("Number of Properties")
sns.despine()
plt.show()
```

```python
sns.set_style('white')
plt.figure(figsize=(8,6))
plt.hist(df_real_estate['Year of sale'], bins=14, color='#108A99')
plt.title('Distribution of Real Estate Years of sale', fontsize=14, fontweight='bold')
# plt.xticks()
plt.xlabel("Year")
plt.ylabel("Number of Properties")
sns.despine()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

df_real_estate1 = pd.read_csv('histogram_data.csv')
df_real_estate1
```
