## Libraries

```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default

from matplotlib.ticker import PercentFormatter #converts values into percentage format
```

## Bar Chart

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
df_used_cars
```

```python
plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"])
plt.show()
```

```python
# how to fix the overlapping labels
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"])
plt.show()
```

```python
# another way to improve visibility is to rotate the labels

plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"])
plt.xticks(rotation=45)
plt.show()
```

```python
# choosing a color
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"], color="rg")
plt.show()
# rgbwymc = red, green, blue, white, yellow, magenta, cyan
```

```python
# to integrate a title
plt.figure(figsize=(9,6))
plt.bar(x = df_used_cars["Brand"], height=df_used_cars["Cars Listings"], color="midnightblue")
plt.xticks(rotation=45, fontsize=13)
plt.yticks(fontsize=13)
plt.title("Cars Listings By Brand", fontsize=16, fontweight="bold")
plt.ylabel("Number of Listings", fontsize=13)
plt.show()
```

```python
# how to save and export image
plt.savefig("used_cars_bar.png")
```

## Pie Chart

```python
# disclaimer: pie charts are considered a bad choice for data viz

# load the dataset using pandas dataframe
df_fuel_engine_type = pd.read_csv("pie_chart_data.csv")

# display the contents of the dataframe
df_fuel_engine_type
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(10,8))
plt.pie(df_fuel_engine_type['Number of Cars'], labels=df_fuel_engine_type['Engine Fuel Type'].values,
                            autopct='%.2f%%',
                            textprops={'size':'x-large',
                                        'fontweight': 'bold',
                                        'rotation': '30',
                                        'color': 'w'})
plt.legend()
plt.title("Cars Engine Fuel Type", fontsize=18, fontweight='bold')
plt.show()
```
