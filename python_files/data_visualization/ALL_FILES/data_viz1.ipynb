## Stacked Area Chart

```python
# an area chart is used to display quantitative data for three or more numerical values
df_fuel_engine_type = pd.read_csv('stacked_area_chart_data.csv')
df_fuel_engine_type
```

```python
# define colors(4)
colors = ['#011638', '#7E2987', 'EF2026']
# define labels(4)
labels = ['Diesel', 'Petrol', 'Gas']
# using seaborn, create a white background
sns.set_style('white')
# increase the size
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_type['Year'],
                df_fuel_engine_type['Diesel'],
                df_fuel_engine_type['Petrol'],
                df_fuel_engine_type['Gas'],
                colors = colors,
                edgecolor = 'none',
                
                )
# include the years and make sure the labels don't overlap
plt.xticks(df_fuel_engine_type['Year'], rotation=45)
# include the labels
plt.legend(labels = labels, loc='upper left')
# create a Y-AXIS label
plt.ylabel('Number of Cars', fontsize=13, fontweight='bold')
# include title
plt.title('Popularity of Engine Fuel Types (1982-2016)', fontsize=14, fontweight='bold')
# despine allows us to remove the borders
sns.despine(left=True, top=True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()


df_ad_exp = pd.read_csv("scatter_plot_ii.csv")
df_ad_exp
```

```python
sns.set(rc={'figure.figsize': (9,6)})
sns.regplot(x='Budget', 
            y='Sales', 
            data=df_ad_exp,
            scatter_kws={'color': 'k'}, 
            line_kws={'color': 'red'})
plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, fontweight='bold')

plt.show()
```

```python
sns.lmplot(x='Budget', 
            y='Sales', 
            data=df_ad_exp,
            height=8, # height parameter controls the height and width
            scatter_kws={'color': 'k'}, 
            line_kws={'color': 'red'})

plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize=14, fontweight='bold')
plt.savefig('Effect of Ad Expenditure on Sales.png')

plt.show()
```

## Bar and Line Chart

```python
df_kd_nuggets = pd.read_csv("bar_line_chart_data.csv")
df_kd_nuggets
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.bar(df_kd_nuggets["Year"], df_kd_nuggets["Participants"], color="k")
ax.set_ylabel("Number of Participants", weight="bold")
ax.tick_params(axis="y", width=2, labelsize="large")
ax1 = ax.twinx()

ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0))

ax1.plot(df_kd_nuggets["Year"], df_kd_nuggets["Python Users"], color="#B60000", marker="D")
ax1.set_ylabel("Python Users", color="#B60000", weight="bold")
ax1.tick_params(axis='y', colors="#B60000", width=2, labelsize="large")
ax.set_title("KD Nuggets Survey Python Users (2012-2019)", fontsize="14", weight="bold")

plt.show()
```
