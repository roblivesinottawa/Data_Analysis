```python
import requests
import json
```

```python
# base url
base_url = "https://itunes.apple.com/search"
```

```python
# make request
params = {"term": "2pac", "country": "us", "limit": 200}
response = requests.get(base_url, params=params)
response.status_code
```

```python
# store response in json
info = response.json()
```

```python
# check if response is valid
print(json.dumps(info, indent=4))
```

```python
# check the first result
print(json.dumps(info['results'][0], indent=4))
```

```python
# where the name is located
info['results'][0]['trackName']
```

```python
# where the release date is located
info['results'][0]['releaseDate']
```

```python
# loop through the results and print their track names
[result['trackName'] for result in info['results']]
```

```python
# loop through the results and print their release dates
[result['releaseDate'] for result in info['results']]
```
